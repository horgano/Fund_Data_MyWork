# Binomial Distribution

https://numpy.org/doc/stable/reference/random/geberated/numpy.random.binomial.html

numpy.random.binomial  random.binomial(n, p, size=None) n=no of parameters, p Probability, n>=0, p is interval [0,1]


```python
# Numerical arrays and random numbers
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Flipping  a singel fair coin one hundred times
rand_coins = np.random.binomial(1, 0.5, 100)

# Show array
rand_coins
```

```python
# Create an empty plot
fig, ax = plt.subplots(figsize=(12, 4))

# Count unique values in array.
x, counts = np.unique(rand_coins, return_counts=True)

# Create a bar chart of rand_coins.
ax.bar(x, counts)
```

#  Flipping a single (possibly biased) coin

```python

# The bernouli distribution is just the binomial with the first parameter as 1

def bernouli(p, size=None):
    return np.random.binomial(1, p, size)
```

```python
# Flip 100 fair coins
bernouli(0.5, 100)
```

```python
rand_coins = bernouli (0.7, 100)
```

```python
# Create an empty plot
fig, ax = plt.subplots(figsize=(12, 4))

# Count unique values in array.
x, counts = np.unique(rand_coins, return_counts=True)

# Create a bar chart of rand_coins.
ax.bar(x, counts)
```

# Rolling Dice


```python
# Default random number generator
rng = np.random.default_rng()
```

```python
# Roll a single dice 1-6(7 not included)
rng.integers(1, 7)
```

```python
# Roll a single dice
single_rolls = rng.integers(1, 7, 10)
```

```python
single_rolls
```

```python
# Count the unique values
x, counts = np.unique(single_rolls, return_counts=True)
```

```python
# Create an empty plot
fig, ax = plt.subplots(figsize=(8,4))

# Create a bar chart of rand_coins.
ax.bar(x, counts)

# Make the x axis look sensible
ax.set_xticks(x);
```

```python
# Roll 2 dice
double_rolls = rng.integers(1, 7, (100000, 2))

# Show a few
double_rolls
```

```python
# Sum the 2 dice
sum_double_rolls = double_rolls.sum(axis=1)

# Show a few
sum_double_rolls
```

```python
# Count the unique values
x, counts = np.unique(sum_double_rolls, return_counts=True)
```

```python
# Create an empty plot
fig, ax = plt.subplots(figsize=(8,4))

# Create a bar chart of rand_coins.
ax.bar(x, counts)

# Make the x axis look sensible
ax.set_xticks(x);
```

# Plotting the individual dice rolls together

```python
# A dirty hack to see both numbers
double_rolls[:,0] * 10 + double_rolls[:,1]
```

```python
# Store the dirty hack
indiv_double_roll = double_rolls[:,0] * 10 + double_rolls[:,1]

# Show
indiv_double_roll
```

```python
# Count the unique values
x, counts = np.unique(indiv_double_roll, return_counts=True)
```

```python
# Create an empty plot
fig, ax = plt.subplots(figsize=(12,6))

# Create a bar chart.
ax.bar(x, counts)

# Make the x axis look sensible
ax.set_xticks(x)

ax.set_xticklabels(x, rotation=90);
```
