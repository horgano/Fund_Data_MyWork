```python
# PLots
import matplotlib.pyplot as plt
```

```python
# Numerical arrays
import numpy as np
```

f(x) = x^2 - 4x + 4

```python
def f(x):
    return x**2 +4*x +4
```

```python
x = np.array([50.0, 70.0, 100.0, 105.0, 123.45])
x
```

```python
y = f(x)
```

```python
y
```

```python
# Create an empty plot
fig, ax = plt.subplots ()

# Plot x vs y
ax.plot(x,y)
```

```python
x = np.array([-3, -2, -1, 0, 1, 2, 3])
```

```python
y = f(x)
```

```python
fig, ax = plt.subplots ()
ax.plot(x,y)
ax.plot(x,y,'.')
```
